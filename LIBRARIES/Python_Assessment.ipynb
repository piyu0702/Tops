```python
# 1) Import pandas and read in the banklist.csv file into a dataframe called banks.

import pandas as pd
banks = pd.read_csv("/content/banklist.csv")
```

```python
# 2) Show the head of the dataframe.
print(banks.head())
```

```python
# 3) What are the column names?
print(banks.columns)

```

```python
# 4) How many States (ST) are represented in this data set?
print("Number of unique states:", banks['ST'].nunique())
```

```python
# 5) Get a list or array of all the states in the data set.
print("States list:", banks['ST'].unique())
```

```python
# 6) What are the top 5 states with the most failed banks?
top_states = banks['ST'].value_counts().head(5)
print("Top 5 states with the most failed banks:")
print(top_states)
```

```python
# 7) What are the top 5 acquiring institutions?
top_institutions = banks['Acquiring Institution'].value_counts().head(5)
print("Top 5 Acquiring Institutions:\n", top_institutions)
```

```python
# 8) How many banks has the State Bank of Texas acquired? How many of them were actually in Texas

bank_count = banks[banks['Acquiring Institution'] == 'State Bank of Texas']
print("Number of banks acquired by State Bank of Texas:", len(bank_count))
in_texas = bank_count[bank_count['ST'] == 'TX']
print("Number of acquired banks actually in Texas:", len(in_texas))
```

```python
# 9) What is the most common city in California for a bank to fail in?
ca_banks = banks[banks['ST'] == 'CA']
most_common_city = ca_banks['City'].value_counts().idxmax()
print("Most common city for bank failure in California:", most_common_city)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
d = pd.read_csv("/content/historical_automobile_sales.csv")
d
```

```python
# Q 1: Develop a Line chart using the functionality of pandas to show how automobile sales fluctuate from year to year.
# Group by Year and sum the Automobile_Sales
yearly_sales = d.groupby('Year')['Automobile_Sales'].sum().reset_index()

# Plot
plt.figure(figsize=(10,6))
plt.plot(yearly_sales['Year'], yearly_sales['Automobile_Sales'], marker='o')
plt.title('Yearly Automobile Sales Trend')
plt.xlabel('Year')
plt.ylabel('Total Automobile Sales')
plt.grid(True)
plt.show()
```

```python
"""
Q 2: Plot different lines for categories of vehicle type and analyze the trend
to answer the question Is there a noticeable difference in sales trends
between different vehicle types during recession periods?
"""
# Group by Year and Vehicle_Type and get mean sales
category_sales = d.groupby(['Year', 'Vehicle_Type'])['Automobile_Sales'].mean().reset_index()

# Plot different lines per Vehicle_Type
plt.figure(figsize=(12,7))
for v_type in category_sales['Vehicle_Type'].unique():
    subset = category_sales[category_sales['Vehicle_Type'] == v_type]
    plt.plot(subset['Year'], subset['Automobile_Sales'], label=v_type)

plt.title('Vehicle Type Sales Trends Over Years')
plt.xlabel('Year')
plt.ylabel('Average Automobile Sales')
plt.legend()
plt.grid(True)
plt.show()
```

```python
"""
Q 3: Use the functionality of Seaborn Library to create a visualization to compare
the sales trend per vehicle type for a recession period with a non- recession
period.
"""

# Make explicit copies to avoid the warning
d_rec = d[d['Recession'] == 1].copy()
d_nonrec = d[d['Recession'] == 0].copy()

# Add a new column to label the period
d_rec['Period'] = 'Recession'
d_nonrec['Period'] = 'Non-Recession'

# Combine both
combined = pd.concat([d_rec, d_nonrec])

# Group by Year, Vehicle_Type, and Period
trend_data = combined.groupby(['Year', 'Vehicle_Type', 'Period'])['Automobile_Sales'].mean().reset_index()

# Plot using Seaborn
plt.figure(figsize=(12,8))
sns.lineplot(data=trend_data, x='Year', y='Automobile_Sales', hue='Vehicle_Type', style='Period', markers=True)
plt.title('Sales Trend per Vehicle Type: Recession vs Non-Recession')
plt.xlabel('Year')
plt.ylabel('Average Automobile Sales')
plt.grid(True)
plt.legend(title='Vehicle Type / Period')
plt.show()
```

```python
# Q 4: Now you want to compare the sales of different vehicle types during a recession and a non-recession period
# Group and calculate average sales
bar_data = d.groupby(['Recession', 'Vehicle_Type'])['Automobile_Sales'].mean().reset_index()

# Map 0 and 1 to readable text
bar_data['Recession'] = bar_data['Recession'].map({1: 'Recession', 0: 'Non-Recession'})

# Plot
plt.figure(figsize=(12,6))
sns.barplot(x='Vehicle_Type', y='Automobile_Sales', hue='Recession', data=bar_data)
plt.title('Average Vehicle Sales: Recession vs Non-Recession')
plt.xlabel('Vehicle Type')
plt.ylabel('Average Automobile Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
